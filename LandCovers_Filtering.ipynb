# Get Land Cover Data
Here you are importing get_api_data which is the GLOBE Observer API data. 
We will be using the landcover_protocol (and you are already familiar with this protocol from the API demonstration!

You can choose where you would like your end_date to be.

```python
from go_utils import get_api_data
from go_utils.constants import landcover_protocol

lc_df = get_api_data(landcover_protocol, end_date="2022-03-30")
```

```python
lc_df
```

# Filter data

We will now use a qa filter method to get all records that have all classifications filled out and all photos taken. 

With this portion there are a couple of options. 
All 6 directional photos (true/false)
All 4 oblique images labeled (true/false)
If you set the option to (false/false) you will get all the data (26,000+ records)

Here we are setting it to False, False. To check if this is correct. you can look at how many rows are filtered out, and this is correct at 26,000+ records!

```python
from go_utils.lc import qa_filter

filtered_lc_df = qa_filter(lc_df, has_all_classifications=False, has_all_photos=False)
```

```python
filtered_lc_df
```

# Diagnostic Plots

We can then use plotting methods to quickly and visually verify that the filtering worked.

The `Valid Photo Count` graph shows that all entries have all 6 photos, the `SubCompleteness Scores` shows that all classifications are filled out.

```python
%matplotlib inline
from go_utils.lc import diagnostic_plots

diagnostic_plots(filtered_lc_df)
```
