```python
## Here We are importing pandas, numpy, and matplotlib. You need to import these packages so you can use them first!

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
## Here we are assigning are data to mydata to make our lives easier. You can name it whatever you would like!

mydata = pd.read_csv("AdoptAPixel3km2020_GO_LandCover.csv")

mydata
```

```python
# Taking the variable landcoversStandingWater and plotting the data as a pie chart
## This is taking the data within the 1047 observations under the column landcoversStandingWater and seeing how many true and how many false observations there are. 
mydata.	landcoversStandingWater.value_counts(sort=False).plot.pie()
plt.show()
```

```python
## This is giving us the mean, median, mode, and also describes the data as well.

mydata[['elevation']].mean()
mydata[['elevation']].median()
mydata[['elevation']].mode()
mydata[['elevation']].describe()
```

```python
## Here we are seeing the outliers within the data by creatiing this box and whiskers plot
## We are using elevation as the column and comparing this to the classification of landcoversStandingWater
## Why is the second plot happening? This is because it is not a numeric value

mydata.boxplot(column='elevation')
plt.show()
mydata.groupby('elevation').boxplot(column='landcoversStandingWater')
plt.show()
```

```python
x=mydata[['elevation']]
type(x)  # Note: x has the same type as mydata, which is a DataFrame
x.sort_values(by=['elevation']) #note, doesn't operate on x
x  # Note: Not sorted!
y=x.sort_values(by=['elevation']) #save a sorted copy if desired
y  # Now sorted!  See it and believe it!
```
